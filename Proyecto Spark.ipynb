3 ¿Cuál es el país que más veces ocupó el primer lugar en todos los años?

Rta: Denmark

```
%scala
import org.apache.spark.sql.functions.{col, max,row_number,count}
import org.apache.spark.sql.expressions.Window

val world_happiness_report= spark.read.format("csv").option("inferSchema",true).option("header",true).load("/FileStore/tables/world_happiness_report.csv")
val windowSpec = Window.partitionBy(col("year")).orderBy(col("Life Ladder").desc)
val rowNumber = row_number().over(windowSpec)

val firstPlaceYear = world_happiness_report.select(col("Country name"), col("Life Ladder"), col("year")).withColumn("top_country", rowNumber).where(col("top_country") === 1)

val countFirstYear = firstPlaceYear.groupBy(col("Country name"),col("top_country")).agg(count("*").alias("country_count")).orderBy(col("country_count").desc)

display(countFirstYear)
```
